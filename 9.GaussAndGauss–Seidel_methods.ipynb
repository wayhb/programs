```python
import math
import pandas as pd
import numpy as np
def mat(a, f, n): #Заполнение матрицы 
    
    for i in range(n):
        for j in range(n):
            if(i==(n-1)):
                a[i][j]=2
            elif(i == j): 
                a[i][j] = i+11
            elif((i==(j+1)) or (i==(j-1))): 
                a[i][j] = 1.
            else: 
                a[i][j] = 0.
        f[i] = i/n 
    f[0]=1
    a[n-1][0] = 1
    a[n-1][n-1] = 1
def norm(a, n):
    norm = 0
    for i in range(n):
        norm += a[i]*a[i]
    return math.sqrt(norm)
def eig(a, n, eps): #Функция поиска максимального
    x = [0] * n #собственного значения
    x[0] = 1 #степенным методом 
    y = [0] * n
    lc = 0.
    ln = 0.
    while True:
        lc = ln
        for i in range(n):
            y[i] = 0.
            for j in range(n):
                y[i] += a[i][j]*x[j]

        ln = 0.
        for i in range(n):
            ln += x[i]*y[i]/pow(norm(x, n),2)

        for i in range(n):
            x[i] = y[i]/norm(y, n)
        if(abs(ln - lc) < eps):
            break
    return ln

def obr(a, n): #Поиск обратной матрицы
    b = [0] * n
    for i in range(n): 
        b[i] = [0] * n
        b[i][i] = 1
    ab = [0] * n
    for i in range(n): 
        ab[i] = [0] * n
    for i in range(n):
        ab[i] = [0] * 2*n

    for i in range(n):
        for j in range(n):
            ab[i][j] = a[i][j]
            ab[i][j + n] = b[i][j]
#Зануление нижнего левого угла
    for k in range(n):
        for i in range(2*n):
            ab[k][i] = ab[k][i]/a[k][k]

        for i in range(k+1,n):
            K = ab[i][k]/ab[k][k]
            for j in range(2*n):
                ab[i][j] -= ab[k][j]*K

        for i in range(n):
            for j in range(n):
                a[i][j] = ab[i][j]
#Зануление верхнего правого угла
    for k in range(n-1,-1,-1):
        for i in range(2*n - 1,-1,-1):
            ab[k][i] = ab[k][i]/a[k][k]
        
        for i in range(k-1,-1,-1):
            K = ab[i][k]/ab[k][k]
            for j in range(2*n - 1,-1,-1):
                ab[i][j] -= ab[k][j]*K


    for i in range(n):
        for j in range(n):
            a[i][j] = ab[i][j + n]
def seidel(A, b, eps):
    n = len(A)
    x = np.zeros(n)  # zero vector

    converge = False
    while not converge:
        x_new = np.copy(x)
        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i))
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n))
            x_new[i] = (b[i] - s1 - s2) / A[i][i]

        converge = np.sqrt(sum((x_new[i] - x[i]) ** 2 for i in range(n))) <= eps
        x = x_new

    return x

def gauss(a, b, x, m): #Метод Гаусса
    for k in range(1,m):
        for j in range(k,m):
            h = a[j][k-1]/a[k-1][k-1]
            for i in range(m):
                a[j][i] = a[j][i]- h*a[k-1][i]
            b[j] = b[j] - h*b[k-1]
            
        for i in range(m-1,-1,-1):
            x[i] = b[i]/a[i][i]
            for c in range(m-1,i,-1):
                x[i] -= a[i][c]*x[c]/a[i][i]
                
                
n = 100
a = [0] * n
for i in range(n): 
    a[i] = [0] * n
b = [0] * n
r = [0] * n
x = [0] * n
mat(a, b, n)
gauss(a,b,x,n)
print("\nRESHENIE GAUS\n")
for i in range(n):
    print(x[i])
#Вычисление вектора невязки
for i in range(n):
    s = 0
    for j in range(n):
        s += a[i][j]*x[j]
    r[i]= b[i] - s
print("NORMA VECTORA NEVIAZKI ", norm(r, n))

print("\nRESHENIE guas-z\n")
seidel(a, b, 1e-6)
for i in range(n):
    print(x[i])
mat(a, b, n)

#Вычисление вектора невязки
for i in range(n):
    s = 0
    for j in range(n):
        s += a[i][j]*x[j]
    r[i]= b[i] - s
print("NORMA VECTORA NEVIAZKI ", norm(r, n))

mu = 1 #Число обусловленности
ax_norm = 0
for i in range(n):
    for j in range(n):
        ax_norm += a[i][j]*x[j] ** 2
mu *= math.sqrt(ax_norm)/pow(norm(x, n), 2)

inv_ax_norm = 0
obr(a, n)
for i in range(n):
    for j in range(n):
        inv_ax_norm += pow(a[i][j]*x[j], 2)
    mu *= math.sqrt(inv_ax_norm)
print("\nCHISLO OBUSLOVLENNOSTI ",  mu)

mat(a, b, n)

#Дополнительные задания
l = eig(a, n, 1e-8)
print("\nMAX SOBSTVENNOE ZNACH ",  l)
obr(a, n)
l = eig(a, n, 1e-10)
print("MIN SOBSTVENNOE ZNACH ",  1./l)
```
