```python
import matplotlib.pyplot as plt
import numpy as np
import math 
x= np.arange(-math.pi,math.pi,math.pi/100)
y1= np.array([-2*a/((1+a*a)*(1+a*a)) for a in x])
h=0.001
y2=np.array([((math.atan(a) - math.atan(a-h))/h - (math.atan(a-h) - math.atan(a-2*h))/h)/h for a in x])

plt.subplot(3,1,1)
plt.plot(x,y1)
plt.title("Аналитическая функция (atan(x))'' ")
plt.grid()
plt.subplot(3,1,2)
plt.plot(x,(y2))
plt.title("Численная функция (atan(x))'' ")
plt.grid()
plt.subplot(3,1,3)
plt.plot(x,(y1-y2))
plt.title("Разностная  функция ")
plt.grid()
plt.show()
```
